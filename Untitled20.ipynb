```python
x1 = np.linspace(-np.pi, np.pi, 256)
x2 = np.linspace(-np.pi,np.pi, 256)

y1 = np.cos(x1)
y2 = np.sin(x2)
```

```python
plt.plot(x1, y1, x2, y2)
```

```python
plt.figure(figsize = (10, 6), dpi = 80)

plt.plot(x1, y1)
plt.plot(x2, y2)

plt.show()
```

```python
plt.figure(figsize = (10, 6), dpi = 80)

plt.plot(x1, y1, 'r-.')
plt.plot(x2, y2, 'b--')

plt.show()
```

```python
plt.figure(figsize = (10, 6), dpi = 80)

plt.plot(x1, y1, color = 'purple')
plt.plot(x2, y2, color = 'green')

plt.show()
```

```python
plt.figure(figsize = (10, 6), dpi = 80)

plt.plot(x1, y1, color = 'purple', lw = 2.5)
plt.plot(x2, y2, color = 'green', lw = 2.5)

plt.show()
```

```python
plt.figure(figsize = (10, 6), dpi = 80)

plt.plot(x1, y1, color = 'purple', lw = 2.5, ls = '--')
plt.plot(x2, y2, color = 'green', lw = 2.5, ls = '-.')

plt.show()
```

```python
plt.figure(figsize = (10, 8), dpi = 90)

plt.plot(x1, y1, color = 'purple', lw = 2.5, ls = '--')
plt.plot(x2, y2, color = 'green', lw = 2.5, ls = '-.')

plt.show()
```

```python
plt.figure(figsize = (10, 6), dpi = 80)

plt.plot(x1, y1, color = 'purple', lw = 2.5, ls = '--')
plt.plot(x2, y2, color = 'green', lw = 2.5, ls = '-.')

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$',r'$\pi/2$',r'$\pi$'])
plt.yticks([-1, 0, 1])

plt.show()
```

```python
plt.figure(figsize = (10, 6), dpi = 80)

plt.plot(x1, y1, color = 'purple', lw = 5, ls = '--', label = '$y = cosx$')
plt.plot(x2, y2, color = 'green', lw = 5, ls = '-.', label = '$y = sinx$')

plt.xlabel('X')
plt.ylabel('Y')

plt.title('Graph')

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$',r'$\pi/2$',r'$\pi$'])
plt.yticks([-1, 0, 1])

plt.legend(loc = 0)

plt.show()
```

```python
x = np.array([1,2,3,4,5])

y1 = x**2
y2 = np.exp(x)
```

```python
plt.figure(figsize = (10, 6), dpi = 80)

plt.subplot(1, 2, 1)
plt.plot(x, y1, 'r')

plt.subplot(1,2, 2)
plt.plot(x, y2, 'b-.')
```
